```python
import os
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

# configuration file and config dir
# you can set default parameters on this config file
print(mpl.matplotlib_fname())
print(mpl.get_configdir())

# to change font settings for one file only
font = {'family' : 'Helvetica Neue', #['serif', 'sans-serif', 'cursive', 'fantasy', 'monospace']
        'style'  : 'normal', #['normal', 'italic', 'oblique']
        'weight' : 'normal', #['light', 'normal', 'medium', 'semibold', 'bold', 'heavy', 'black']
        'size'   : 12}
mpl.rc('font', **font)

# 1. Add any needed font (.ttf format) in the install folder of matplotlib
#    On OSX using miniconda /Users/xxxxx/miniconda3/lib/python3.6/site-packages/matplotlib/mpl-data/fonts/ttf/
# 2. Delete fontList.json (or fontList.cache) located at ~/.matplotlib/
# 3. Restart (or reload) matplotlib library to rebuild font-cache with fc-list

# if you want to find out the family name of a font in the folder
path = os.path.expanduser('~/miniconda3/lib/python3.6/site-packages/matplotlib/mpl-data/fonts/ttf/HelveticaNeue.ttf')
print(mpl.font_manager.FontProperties(fname=path).get_name())
```

```python
# See results
t = np.arange(0.0, 2.0, 0.01)
s = 1.0 + np.cos(2*np.pi*t)
plt.plot(t, s)

plt.xlabel('time (s)')
plt.ylabel('Position (m)')
plt.title('Title and axis with a brand new font')
plt.show()
```
